# Mathplotlib Tutorial

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Functional Plot
x = np.linspace(0,5,10)
y = x**2
plt.plot(x,y)
plt.title("Days Squared Chart")
plt.xlabel("Days")
plt.ylabel("Days Squared")
```

```python
#PRint Multiplr Plots
plt.subplot(1,2,1)
plt.plot(x,y,'y')
plt.subplot(1,2,2)
plt.plot(y,x,'k')
```

```python
# Using Figure Objects
fig = plt.figure(figsize = (5,4),dpi=100)
axes = fig.add_axes([0.1,0.1,0.9,0.9])
axes.set_xlabel('Days')
axes.set_ylabel('Days Squared')
axes.set_title('Days Sqaured Charts')
axes.plot(x,y,label='x / x²')
axes.plot(y,x,label= 'x²/x')
axes.legend(loc = 0)

```

```python
axes2 = fig.add_axes([0.45,0.45,0.4,0.3])
axes2.set_xlabel('Days')
axes2.set_ylabel('Days Squared')
axes2.set_title('small Days Sqaured Chart')
axes2.plot(x,y,'r')
axes2.text(0,40, "Message")
```
